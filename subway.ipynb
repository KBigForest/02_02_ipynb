## 실습: 출근 시간대 지하철 이용현황

```python
import csv
import matplotlib.pyplot as plt
with open('subwaytime.csv',encoding = 'utf-8-sig') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    result= []
    total_number = 0
    max_num = -1
    max_station = ''
    for row in data:
        row[4:] = map(int, row[4:])
        row_sum = sum(row[10:15:2])
        result.append(row_sum)
        if row_sum > max_num:
            max_num = row_sum
            max_station = row[3]+ '('+ row[1] + ')'
            
print('최대 승차 인원역: {0} {1:,}'.format(max_station, max_num))
result.sort(reverse= True)
plt.figure(dpi = 100)
plt.bar(range(10), result[0:10])
# plt.bar(range(len(result)), result)
plt.xticks(range(10))
plt.show()
    
```

## 시간대별 가장 많이 승차하는 역 정보 분석

```python
import csv
import matplotlib.pyplot as plt

with open('subwaytime.csv',encoding = 'utf-8-sig') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    max = [0] * 23
    max_station =['']*23
    xtick_list =[]
    
    for i in range(4, 27):
        n = i % 24
        xtick_list.append(str(n))
    for row in data:
        row[4:] = map(int, row[4:])
        for j in range(23):
            a = row[j * 2 + 4]
            if a > max[j]:
                max[j] = a
                max_station[j] = xtick_list[j]+'시: '+row[3]
    for i in range(len(max)):
        print('{0}: {1:,}'.format(max_station[i], max[i]), end= ', ')
        if (i+1) % 4 ==0:
            print()
if (platform.system() == 'Windows'):
    plt.rcParams['font.family'] = 'D2Coding'

                      
plt.figure(dpi = 100)
plt.bar(range(23),max)
plt.xticks(range(23), max_station, rotation = 80)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
xtick_list=[]
with open('subwaytime.csv',encoding = 'utf-8-sig') as f:
    data = csv.reader(f)
    next(data)
    next(data)
    s_in = [0] * 24
    s_out= [0] * 24
    
    for i in range(4, 28):
        n = i % 24
        xtick_list.append(str(n))
    for row in data:
        row[4:] = map(int, row[4:])
        for j in range(24):
            s_in[j] += row[4+j*2]
            s_out[j] += row[5+j*2]

if (platform.system() == 'Windows'):
    plt.rcParams['font.family'] = 'D2Coding'

plt.figure(dpi = 100)
plt.title('지하철 시간대별 승하차 인원 추이',size= 16)
plt.grid(linestyle= ':')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()

plt.xticks(range(24), labels = xtick_list)
plt.xlabel('시간')
plt.ylabel('인원(천만명)')
plt.show()
```

## lambda와 operator를 사용한 Dictionary 정렬
### operator 모듈과 lambda 사용 정렬

```python
import operator as op

names = {'Mary':10999, 'Sams':2111,	'Aimy':9778, 'Tom':20245, 'Michale':27115, 'Bob':5887, 'Kelly':7855}


res = sorted(names.items(), key= (lambda x: x[0]))
print(res)

res = sorted(names.items(), key= (lambda x: x[1]), reverse =True)
print(res)
```

```python
import operator as op
sorted_x1 = sorted(names.items(), key = op.itemgetter(0))
print(sorted_x1)
sorted_x2 = sorted(names.items(), key = op.itemgetter(1), reverse= True)
print(sorted_x2)
```
