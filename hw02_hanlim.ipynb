```python
# 각 노선 별 최대 하차 인원 막대그래프로 표시하고, 하차인원 출력
import pandas as pd
import matplotlib.pyplot as plt
import platform as pf

df_1 = pd.read_csv('subwaytime.csv', encoding='utf-8-sig',header =[0,1])


    
```

```python
x= ['1호선', '2호선', '3호선', '4호선', '5호선', '6호선', '7호선']
df_0709_out = df_1[[ ('지하철역', 'Unnamed: 3_level_1'),('호선명', 'Unnamed: 1_level_1'),('Unnamed: 11_level_0','하차'),('Unnamed: 13_level_0','하차')]]
df_0709_out[df_0709_out[('호선명', 'Unnamed: 1_level_1')].isin(x)]
row_sum_list = df_0709_out.sum(axis = 1,numeric_only=True)
df_0709_out[('승객', '7~9시')]= row_sum_list
df_0709_out.drop([('Unnamed: 11_level_0','하차'),('Unnamed: 13_level_0','하차')],axis = 1,inplace=True)
df_0709_out
```

```python

max_passenger_list = []
max_station_list = []
max_line_list= []
line_station_list = []
for i in x:
    max_station = df_0709_out[df_0709_out[('호선명', 'Unnamed: 1_level_1')]==i].max()[0]
    max_line = df_0709_out[df_0709_out[('호선명', 'Unnamed: 1_level_1')]==i].max()[1]
    max_passenger = df_0709_out[df_0709_out[('호선명', 'Unnamed: 1_level_1')]==i].max()[2]
    line_station = max_line +' ' + max_station
    max_station_list.append(max_station)
    max_line_list.append(max_line)
    max_passenger_list.append(max_passenger)
    line_station_list.append(line_station)

    
for i in range(7):
    print('출근 시간대 {0} 최대 하차역 : {1}, 하차인원: {2:,}명'.format(max_line_list[i], max_station_list[i], max_passenger_list[i]))
if (pf.system() == 'Windows'):
    plt.rcParams['font.family'] = 'D2Coding'
plt.figure(dpi = 100)
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역')
plt.bar(line_station_list, max_passenger_list)
plt.xticks(line_station_list, rotation = 80)
plt.show()
```

```python
# 각 노선 별 최대 하차 인원 막대그래프로 표시하고, 하차인원 출력
import pandas as pd
import matplotlib.pyplot as plt
import platform as pf

df_1 = pd.read_csv('subwaytime.csv', encoding='utf-8-sig',header =[0,1])
x= ['1호선', '2호선', '3호선', '4호선', '5호선', '6호선', '7호선']
df_0709_out = df_1[[ ('지하철역', 'Unnamed: 3_level_1'),('호선명', 'Unnamed: 1_level_1'),('Unnamed: 11_level_0','하차'),('Unnamed: 13_level_0','하차')]]
df_0709_out[df_0709_out[('호선명', 'Unnamed: 1_level_1')].isin(x)]
row_sum_list = df_0709_out.sum(axis = 1,numeric_only=True)
df_0709_out[('승객', '7~9시')]= row_sum_list
df_0709_out.drop([('Unnamed: 11_level_0','하차'),('Unnamed: 13_level_0','하차')],axis = 1,inplace=True)
df_0709_out = df_0709_out.groupby(('지하철역', 'Unnamed: 3_level_1')).sum()
result_df = df_0709_out.sort_values(('승객', '7~9시'),ascending=False).head()
```

```python
if (pf.system() == 'Windows'):
    plt.rcParams['font.family'] = 'D2Coding'

station_list = []
passenger_num_list = []
for i in range(0,5):
    print('{0}역: {1:,}명'.format(result_df.iloc[i].name,result_df.iloc[i][0]))
    station_list.append(result_df.iloc[i].name)
    passenger_num_list.append(result_df.iloc[i][0])
plt.figure(dpi = 100)
plt.title('서울 지하철 출근시간대 하차 인원 비교(7:00~8:59)')
plt.bar(station_list, passenger_num_list, color = 'blue')
plt.show()
```
